```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plot
 
```

```python
iris = pd.read_csv("Iris.csv")
```

```python
iris.head()
```

```python
iris.describe()
```

```python
iris.isnull().sum()
```

```python
# PDF

counts, bins = np.histogram(iris["SepalLengthCm"],bins = 10, density= False)
pdf = counts / (sum(counts))

# plot.gca().legend(("pdf"))

plot.title("PDF")
plot.xlabel("SEPAL LENGTH")
plot.ylabel("PERCENTAGE")
plot.plot(bins[1:],pdf)
```

```python
# CDF

counts , bins = np.histogram(iris["SepalLengthCm"],bins = 10,density=True)
pdf = counts /(sum(counts))

cdf = np.cumsum(pdf)

plot.title("CDF")
plot.xlabel("SEPAL LENGTH")
plot.ylabel("PERCENTAGE")
plot.plot(bins[1:],cdf)
```

```python
pmf,bins = np.histogram(iris["SepalLengthCm"],bins = range(0,11),density=True)

plot.title("PMF")
plot.xlabel("SEPAL Length")
plot.ylabel("PERCENTAGE")
plot.plot(bins[1:],pmf)
```
