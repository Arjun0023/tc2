```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plot

from scipy.stats import uniform,norm,expon,binom,poisson
```

```python
iris = pd.read_csv("Iris.csv")
```

```python
np.random.seed(42)

data = iris['SepalLengthCm'].values
sample_size = len(data)
```

```python
# Uniform Distribution

uniform = np.random.uniform(low=min(data),high=max(data),size=sample_size)

plot.hist(uniform, bins = 20,  density=True)
```

```python
# Normal Distribution

mean,std_dev = np.mean(data), np.std(data)

normal = np.random.normal(loc = mean, scale=std_dev,size=sample_size)

plot.hist(normal, bins = 20, density = True)
```

```python
# Exponential Distribution

exponential = np.random.exponential(scale = np.mean(data),size= sample_size)

plot.hist(exponential, bins = 20, density=True)
```

```python
# Binomial Distribution

n = 10  #number of Trial
p = 0.5  #probability of Success

binomial = np.random.binomial(n, p, size= sample_size)

plot.hist(binomial, bins = np.arange(0, n+2) - 0.5 , density= True)
```

```python
# Poisson Distribution

lambda_parameter = 3    # poisson Parameter

poisson = np.random.poisson(lambda_parameter, size= sample_size)

plot.hist(poisson, bins = 20, density= True)
```
