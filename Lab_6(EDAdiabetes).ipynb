```python
#8. do a statistical summary of the data
df = df[df['Insulin']!=0]
df = df[df['BloodPressure']!=0]
df = df[df['SkinThickness']!=0]
df = df[df['BMI']!=0]
df = df[df['Glucose']!=0]
```

```python
df.describe
```

```python
#9. Display Scatterplots , Heatmaps Box plots and Bar Graphs .

sns.scatterplot(data=df, x="Insulin", y="Glucose", hue="Outcome" )
plt.show()
```

```python
corr = df.corr(method='spearman')
sns.heatmap(corr, annot=True)
```

```python
sns.FacetGrid(df, hue="Outcome", ) \
   .map(sns.distplot, "Glucose") \
   .add_legend();
plt.show();
```

```python
sns.boxplot(x='Outcome',y='Glucose', data=df)
plt.show()
```
